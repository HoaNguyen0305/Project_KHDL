```python
import selenium
from time import sleep
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm
import csv
from selenium import webdriver 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException,TimeoutException, WebDriverException
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from random import randint
from time import sleep
```

```python
def collect_data(x, end):


    list_room=[]
    list_area=[]
    list_price=[]
    list_price_per_m2=[]
    list_wc=[]
    list_address=[]
    while x <= end:

        driver = webdriver.Chrome(ChromeDriverManager().install())
        driver.get('https://batdongsan.com.vn/ban-nha-rieng-tp-hcm/p'+str(x))
        total_height = int(driver.execute_script("return document.body.scrollHeight"))
        for i in range(0, total_height, 300):
            driver.execute_script("window.scrollTo(0, {});".format(i))
            
        houses=driver.find_elements(By.CLASS_NAME, "js__card.js__card-full-web.pr-container.re__card-full")

        # lặp qua từng ngôi nhà và lấy thông tin
        for house in houses:
            # lấy giá nhà
            price = house.find_element(By.CSS_SELECTOR, "span.re__card-config-price.js__card-config-item")
            print("Giá nhà:", price.text)
            list_price.append(price.text)

            #lấy diện tích nhà
            area = house.find_element(By.CSS_SELECTOR, "span.re__card-config-area.js__card-config-item")
            print("Diện tích:", area.text)
            list_area.append(area.text)

            # lấy giá trên 1m2
            try:
                price_per_m2 = house.find_element(By.CSS_SELECTOR, "span.re__card-config-price_per_m2.js__card-config-item")            
                print("giá trên 1m2:", price_per_m2.text)
                list_price_per_m2.append(price_per_m2.text)
            except NoSuchElementException:  #spelling error making this code not work as expected
                list_price_per_m2.append(None)

            #lấy số lượng phòng
            try:
                room = house.find_element(By.CSS_SELECTOR, "span.re__card-config-bedroom.js__card-config-item")
                print("Số lượng phòng ngủ:", room.text)
                list_room.append(room.text)
            except NoSuchElementException:  #spelling error making this code not work as expected
                list_room.append(None)

            #lấy số lượng WC
            try:
                room_wc = house.find_element(By.CSS_SELECTOR, "span.re__card-config-toilet.js__card-config-item")
                print("Số lượng phòng WC:", room.text)
                list_wc.append(room_wc.text)
            except NoSuchElementException:  #spelling error making this code not work as expected
                list_wc.append(None)
           
            # lấy địa chỉ
            address = house.find_element(By.CSS_SELECTOR, "div.re__card-location")
            print("Địa chỉ: ", address.text)
            list_address.append(address.text)
           
        print("next page: " + str(x))
        sleep(randint(1,5))
        x +=1
        
    return list_price,list_room,list_area,list_price_per_m2,list_wc,list_address
```

```python
list_price,list_room,list_area,list_price_per_m2,list_wc,list_address= collect_data(1,50)
```

```python
df = pd.DataFrame({'Address' : list_address,
                    'Area' : list_area,
                    'Price per m2' : list_price_per_m2,
                    'Room':list_room, 
                    'WC':list_wc,
                    'Price':list_price},)
```
